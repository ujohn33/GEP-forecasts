```python
import pandas as pd
import numpy as np
import os
import datetime
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
DATA_DIR = os.path.abspath('')
TARGET_DIR = os.path.dirname(DATA_DIR)+'/Columbia_clean'
```

```python
for i in range(1,28):
    filename = '/Residential_'+str(i)+'.csv'
    temp = pd.read_csv(DATA_DIR+filename)
    print(filename)
    temp.index = pd.to_datetime(temp.date.astype(str) + '.' + temp.hour.astype(str)+':00:00')
    temp = temp.drop(['date', 'hour'], axis=1)
    print(temp.columns)
    temp.to_csv(TARGET_DIR+filename, index=True)
```

```python
for i in range(1,28):
    filename = '/Residential_'+str(i)+'.csv'
    temp = pd.read_csv(TARGET_DIR+filename, index_col=0)
    temp.index = pd.to_datetime(temp.index)
    years = temp.index.year.unique()
    # Prep Colors
    np.random.seed(100)
    mycolors = np.random.choice(list(mpl.colors.XKCD_COLORS.keys()), len(years), replace=False) 
    # Draw Plot
    fig, ax = plt.subplots(figsize=(16,12), dpi= 80)
    for d,year in enumerate(years):
            data_temp=temp.loc[temp.index.year==year, :]
            data_temp.plot(y='energy_kWh', use_index=True, color=mycolors[d], label = year, ax=ax, alpha=0.5)
    # Decoration
    plt.gca().set(ylabel='Consumption [kWh]', xlabel='# measurement')
    plt.yticks(fontsize=12, alpha=.7)
    plt.legend(fontsize=20)
    plt.title("Consumption in residential building "+str(i)+" by year", fontsize=20)
    plt.show()
    fig.savefig(TARGET_DIR+'/Residential_'+str(i)+'.png')   # save the figure to file
    plt.close(fig) 
```
