```python
import pandas as pd
import numpy as np
import os
import datetime
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
DATA_DIR = os.path.abspath('')
TARGET_DIR = os.path.dirname(DATA_DIR)+'/EFC2012_clean'
```

```python
filename = '/Load_history'+'.csv'
temp = pd.read_csv(DATA_DIR+filename)
temp['datetime'] = pd.to_datetime(temp.year.astype(str) + '/' + temp.month.astype(str)+'/'+temp.day.astype(str))
temp
```

```python
temp2 = temp.melt(id_vars=['datetime','zone_id'], 
        value_vars=list(temp.columns[4:-1]),
        var_name='hour', 
        value_name='Consumption') 
```

```python
temp2.head()
```

```python
temp2[temp2['datetime']=='2004-01-01']
```

```python
temp2['hour'] = temp2['hour'].str.replace(r'\D', '')
temp2['Consumption'] = temp2['Consumption'].str.replace(',', '.')
```

```python
temp2.index = pd.to_datetime(temp2.datetime.astype(str) + ' ' + (temp2.hour.astype(int).subtract(1)).astype(str)+':00:00')
```

```python
temp2.sort_index()
```

```python
temp2 = temp2.drop(['datetime', 'hour'], axis=1)
```

```python
zones=temp2['zone_id'].unique().tolist()
for zone in zones:
    joe = temp2.loc[temp2.zone_id==zone]
    joe = joe.drop(['zone_id'], axis=1)
    joe.to_csv(TARGET_DIR+'/zone_'+str(zone)+'.csv')
```

```python
zones=temp2['zone_id'].unique().tolist()
for zone in zones:
    filename = '/zone_'+str(zone)+'.csv'
    temp = pd.read_csv(TARGET_DIR+filename, index_col=0)
    temp.index = pd.to_datetime(temp.index)
    years = temp.index.year.unique()
    #print(years)
    # Prep Colors
    np.random.seed(100)
    mycolors = np.random.choice(list(mpl.colors.XKCD_COLORS.keys()), len(years), replace=False) 
    # Draw Plot
    fig, ax = plt.subplots(figsize=(16,12), dpi= 80)
    for d,year in enumerate(years):
            data_temp=temp.loc[temp.index.year==year, :]
            #print(data_temp['Consumption'])
            data_temp=data_temp.astype(float)
            data_temp.plot(y='Consumption', use_index=True, color=mycolors[d], label = year, ax=ax, alpha=0.5)
    # Decoration
    plt.gca().set(ylabel='Consumption', xlabel='# measurement')
    plt.yticks(fontsize=12, alpha=.7)
    plt.legend(fontsize=20)
    plt.title("Consumption in a utility zone "+str(zone)+" by year", fontsize=20)
    plt.show()
    fig.savefig(TARGET_DIR+'/zone_'+str(zone)+'.png')   # save the figure to file
    plt.close(fig) 
```
