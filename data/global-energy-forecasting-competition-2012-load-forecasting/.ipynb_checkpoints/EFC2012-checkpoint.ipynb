```python
import pandas as pd
import numpy as np
import os
import datetime
```

```python
DATA_DIR = os.path.abspath('')
TARGET_DIR = os.path.dirname(DATA_DIR)+'/EFC2012_clean'
```

```python
filename = '/Load_history'+'.csv'
temp = pd.read_csv(DATA_DIR+filename)
temp['datetime'] = pd.to_datetime(temp.year.astype(str) + '/' + temp.month.astype(str)+'/'+temp.day.astype(str))
temp
```

```python
pd.melt(temp, id_vars=['datetime'], 
        value_vars=list(temp.columns[4:]), # list of days of the week
        var_name='hour', 
        value_name='Consumption')
```

```python
temp2 = temp.melt(id_vars='datetime', 
        value_vars=list(temp.columns[4:-1]),
        var_name='hour', 
        value_name='Consumption') 
```

```python
temp2.head()
```

```python
temp2[temp2['datetime']=='2004-01-01']
```

```python
temp2['hour'] = temp2['hour'].str.replace(r'\D', '')
```

```python
temp2.head()
temp2.dtypes
```

```python
temp2.index = pd.to_datetime(temp2.datetime.astype(str) + ' ' + (temp2.hour.astype(int).subtract(1)).astype(str)+':00:00')
```
