```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
fram = pd.read_csv('results/Dieteren_fram.csv', index_col=0)
MIMO_full = pd.read_csv('results/Dieteren_MIMO-full.csv', index_col=0)
MIMO = pd.read_csv('results/Dieteren_MIMO.csv', index_col=0)
condRNN = pd.read_csv('results/Dieteren_ConditionalRNN.csv', index_col=0)
datasets = [MIMO, condRNN, MIMO_full, fram]
```

```python
condRNN
```

```python
means = []
stds = []
#means.append(35.58681558224254)
#stds.append(0)
for df in datasets:
    means.append(df['rmse'].mean())
    stds.append(df['rmse'].std())
```

```python
means
```

```python
plt.rcParams['figure.figsize'] = (10, 8)
colors=["green", "green", "green", "green"]
labels = ['MIMO', 'ConditionalRNN', 'MIMO-full_data', 'LSTM-framework']
plt.bar(range(len(means)), means, yerr=stds, align='center', alpha=0.5, color=colors)
plt.xticks(ticks=range(len(means)), labels=labels,
           rotation=45)
plt.title('Comparison between model acccuracy for day-ahead load forecast\n(Dieteren data - last 2 months 2016)')
plt.ylabel('RMSE')
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (10, 8)
colors=["green", "green", "green", "green"]
labels = ['MIMO', 'ConditionalRNN', 'MIMO-full_data', 'LSTM-framework']
plt.bar(range(len(means)), means, yerr=stds, align='center', alpha=0.5, color=colors)
plt.xticks(ticks=range(len(means)), labels=labels,
           rotation=45)
plt.title('Comparison between model acccuracy for day-ahead load forecast\n(Dieteren data - last 2 months 2016)')
plt.ylabel('MAPE (%)')
plt.show()
```
