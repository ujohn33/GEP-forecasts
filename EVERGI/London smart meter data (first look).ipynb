```python
halfhourly = pd.read_csv(DATADIR+'./halfhourly_dataset/halfhourly_dataset/block_0.csv')
halfhourly
```

```python
halfhourly.loc[halfhourly['energy(kWh/hh)']=='Null', 'energy(kWh/hh)'] = np.nan
```

```python
halfhourly.index = pd.to_datetime(halfhourly['tstp'])
halfhourly['energy(kWh/hh)'] = halfhourly['energy(kWh/hh)'].str.strip().astype(np.int64)
```

```python
halfhourly['energy(kWh/hh)'] = halfhourly['energy(kWh/hh)'].normalize()
```

```python
halfhourly[halfhourly.isna().any(axis=1)]
```

```python
hourly = halfhourly.groupby('LCLid')['energy(kWh/hh)'].resample('H').sum().to_frame()
```

```python
hourly
```

```python
hourly.to_csv('hourly_all.csv')
```
