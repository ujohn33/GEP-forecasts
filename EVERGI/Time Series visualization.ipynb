```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

data = pd.read_pickle('building_1_full.pkl').loc['01/01/2015 00:00':
                                                 '31/12/2019 23:45']
data.head()
```

```python
data.tail()
```

```python
#fig, ax = plt.subplots()

years = data.iloc[:,:5]

# Prep Colors
np.random.seed(100)
mycolors = np.random.choice(list(mpl.colors.XKCD_COLORS.keys()), len(years.columns), replace=False) 

# Draw Plot
#plt.figure(figsize=(16,12), dpi= 80)
fig, ax = plt.subplots(figsize=(16,12), dpi= 80)
for i,year in enumerate(list(years.columns)):        
        years.plot(y=year, use_index=True, color=mycolors[i], ax=ax, alpha=0.5)
        #plt.text(df.loc[df.year==y, :].shape[0]-.9, df.loc[df.year==y, 'value'][-1:].values[0], y, fontsize=12, color=mycolors[i])

# Decoration
plt.gca().set(ylabel='Consumption [kWh]', xlabel='# measurement')
plt.yticks(fontsize=12, alpha=.7)
plt.title("Yearly Consumption in building 1", fontsize=20)
plt.show()
```
