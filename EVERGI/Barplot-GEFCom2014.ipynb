```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
fram = pd.read_csv('results/GEFCom_fram.csv', index_col=0)
MIMO_full = pd.read_csv('results/GEFCom_MIMO-full.csv', index_col=0)
MIMO = pd.read_csv('results/GEFCom_MIMO.csv', index_col=0)
condRNN = pd.read_csv('results/GEFCom_ConditionalRNN.csv', index_col=0)
datasets = [MIMO, condRNN, MIMO_full, fram]
```

```python
means = []
stds = []
means.append(22.475545630579663)
stds.append(0)
means.append(19.5)
stds.append(0.5)
means.append(19.0)
stds.append(0.2)
means.append(18.0)
stds.append(0.3)
means.append(17.1)
stds.append(0.2)
for df in datasets:
    means.append(df['rmse'].mean())
    stds.append(df['rmse'].std())
```

```python
means
```

```python
plt.rcParams['figure.figsize'] = (10, 8)
colors=['blue', 'black','black','black', 'black',"green", "green", "green", "green"]
labels = ['Baseline\n (same as yesterday)', 'LSTM\n Reference','GRU \nReference','ERNN \nReference','Seq2Seq \nReference','LSTM MIMO', 'ConditionalRNN', 'LSTM MIMO-full_data', 'LSTM-framework']
plt.bar(range(len(means)), means, yerr=stds, align='center', alpha=0.5, color=colors)
plt.xticks(ticks=range(len(means)), labels=labels,
           rotation=45)
plt.title('Comparison between model acccuracy for day-ahead load forecast\n(GEFCOM2014 data 2005-2011)')
plt.show()
```
