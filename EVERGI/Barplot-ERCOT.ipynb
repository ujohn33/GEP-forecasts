```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
fram = pd.read_csv('results/ERCOT_fram.csv', index_col=0)
MIMO_full = pd.read_csv('results/ERCOT_MIMO-full.csv', index_col=0)
MIMO = pd.read_csv('results/ERCOT_MIMO.csv', index_col=0)
condRNN = pd.read_csv('results/ERCOT_ConditionalRNN.csv', index_col=0)
datasets = [MIMO, condRNN, MIMO_full, fram]
```

```python
means = []
stds = []
means.append(6.74308325359656)
stds.append(0)
means.append(5.59)
stds.append(0)
for df in datasets:
    means.append(df['mape'].mean())
    stds.append(df['mape'].std())
```

```python
means
```

```python
plt.rcParams['figure.figsize'] = (10, 8)
colors=['blue', 'black', "green", "green", "green", "green"]
labels = ['Baseline\n (same as yesterday)', 'Result in Paper','MIMO', 'ConditionalRNN', 'MIMO-full_data', 'LSTM-framework']
plt.bar(range(6), means, yerr=stds, align='center', alpha=0.5, color=colors)
plt.xticks(ticks=range(6), labels=labels,
           rotation=45)
plt.title('Comparison between model acccuracy for day-ahead load forecast\n(ERCOT data 2012-2015)')
plt.ylabel('MAPE [%]')
plt.show()
```
