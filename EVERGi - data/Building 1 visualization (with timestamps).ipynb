```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
data2014 = pd.read_pickle('Building-1-Roulart Media Group/541448810000124194_01012014_31122014.pkl')
data2015 = pd.read_pickle('Building-1-Roulart Media Group/541448810000124194_01012015_31122015.pkl')
data2016 = pd.read_pickle('Building-1-Roulart Media Group/541448810000124194_01012016_31122016.pkl')
data2017 = pd.read_pickle('Building-1-Roulart Media Group/541448810000124194_01012014_31122014.pkl')
data2018 = pd.read_pickle('Building-1-Roulart Media Group/541448810000124194_01012014_31122014.pkl')
data2019 = pd.read_pickle('Building-1-Roulart Media Group/541448810000124194_01012014_31122014.pkl')
```

```python
frames = [data2014,data2015,data2016,data2017,data2018,data2019]
data = pd.concat(frames)
```

```python
data.index = pd.to_datetime(data.index)
data['year'] = data.index.year
data['month'] = data.index.month
data['day'] = data.index.day
data['week'] = data.index.week
data['day of year'] = data.index.dayofyear
data['day of week'] = data.index.dayofweek
data['2014-07-03':]
```

```python
years = data['year'].unique()

# Prep Colors
np.random.seed(100)
mycolors = np.random.choice(list(mpl.colors.XKCD_COLORS.keys()), len(years), replace=False) 

# Draw Plot
fig, ax = plt.subplots(figsize=(16,12), dpi= 80)
for i,year in enumerate(years):
        data_temp=data.loc[data.year==year, :]
        data_temp.plot(y='Valeur', use_index=True, color=mycolors[i], ax=ax, alpha=0.5)

# Decoration
plt.gca().set(ylabel='Consumption [kWh]', xlabel='# measurement')
plt.yticks(fontsize=12, alpha=.7)
plt.title("Consumption in building 1 by year", fontsize=20)
plt.show()
```

```python
years = data['year'].unique()

# Prep Colors
np.random.seed(100)
mycolors = np.random.choice(list(mpl.colors.XKCD_COLORS.keys()), len(years), replace=False) 

# Draw Plot
fig, ax = plt.subplots(figsize=(16,12), dpi= 80)
for i,year in enumerate(years):
        data_temp=data.loc[data.year==year, :]
        plt.scatter(x='day of year', y='Valeur', data = data_temp, s=0.1, color=mycolors[i], alpha=0.5, label=year)

# Decoration
plt.gca().set(ylabel='Consumption [kWh]', xlabel='day in the year')
plt.yticks(fontsize=12, alpha=.7)
plt.title("Yearly Consumption in building 1", fontsize=20)
plt.show()
```

```python
# Draw Plot
fig, axes = plt.subplots(1, 4, figsize=(20,7), dpi= 80)
sns.boxplot(x='year', y='Valeur', data=data, ax=axes[0])
sns.boxplot(x='month', y='Valeur', data=data, ax=axes[1])
sns.boxplot(x='day of week', y='Valeur', data=data, ax=axes[2])
sns.boxplot(x=data.index.time, y='Valeur', data=data, ax=axes[3])


# Set Title
axes[0].set_title('Year-wise Box Plot\n(The Trend)', fontsize=18); 
axes[1].set_title('Month-wise Box Plot\n(The Seasonality)', fontsize=18)
axes[2].set_title('Day-of-week Box Plot\n(The Seasonality)', fontsize=18)
axes[3].set_title('Time of the day Box Plot\n(The Seasonality)', fontsize=18)

plt.show()
```
